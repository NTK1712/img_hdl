```python
import numpy as np

def gaussian_kernel(size, sigma):
    kernel = np.fromfunction(
        lambda x, y: (1 / (2 * np.pi * sigma ** 2)) * np.exp(- ((x - (size-1)/2)**2 + (y - (size-1)/2)**2) / (2 * sigma ** 2)),
        (size, size)
    )
    return kernel
kernel_size = 3
sigma = 1.5
# Tạo kernel bộ lọc Gaussian
gaussian_filter = gaussian_kernel(kernel_size, sigma)

print("Kernel bộ lọc Gaussian:")
print(gaussian_filter)
```

```python
#txt_to_image

from PIL import Image

def hex_to_dec(file_path, width, height, output_file_path):

    with open(file_path, "r") as file:
        lines = file.readlines()

    content_array = []
    for line in lines:
        if '//' in line:
            continue
        content_array.append(line.strip())

    binary_list = []
    for i in range(width*height):
        if content_array[i]:
            binary_element = int(content_array[i], 16)
            binary_list.append(binary_element)
        else:
            continue

    with open(output_file_path, "w") as output_file:
        for binary_element in binary_list:
            output_file.write(str(binary_element)+"\n")



def txt_to_image(txt_file_path, image_width, image_height, output_image_path):
    with open(txt_file_path, 'r') as file:
        content = file.readlines()

    content_array = []
    for line in content:
        content_array.append(int(line.strip()))

    img = Image.new('L', (image_width, image_height))  

    for y in range(image_height):
        for x in range(image_width):
            pixel_value = content_array[y * image_width + x]
            img.putpixel((x, y), pixel_value)

    img.save(output_image_path)


hex_to_dec("./output.txt",300,400, "./temp.txt")
txt_to_image("./temp.txt", 300, 400, "./output.png")
```

```python
#txt_to_rgb_image

from PIL import Image

def hex_to_dec(file_path, width, height, output_file_path):

    with open(file_path, "r") as file:
        lines = file.readlines()

    content_array = []
    for line in lines:
        if '//' in line:
            continue
        content_array.append(line.strip())

    binary_list = []
    for i in range(width*height*3):
        if content_array[i]:
            binary_element = int(content_array[i], 16)
            binary_list.append(binary_element)
        else:
            continue

    with open(output_file_path, "w") as output_file:
        for binary_element in binary_list:
            output_file.write(str(binary_element)+"\n")



def txt_to_image(txt_file_path, image_width, image_height, output_image_path):
    with open(txt_file_path, 'r') as file:
        content = file.readlines()

    content_array = []
    for line in content:
         content_array.append(line.strip())

    img = Image.new("RGB", (image_width, image_height))

    index = 0
    for y in range(image_height):
        for x in range(image_width):

            red = int(content_array[index])
            green = int(content_array[index + 1])
            blue = int(content_array[index + 2])

            img.putpixel((x, y), (red, green, blue))

            index += 3

    img.save(output_image_path)


hex_to_dec("./output.txt", 300, 400,"./temp.txt")
txt_to_image("./temp.txt", 300, 400, "./output.png")
```
