```python
import numpy as np

def gaussian_kernel(size, sigma):
    kernel = np.fromfunction(
        lambda x, y: (1 / (2 * np.pi * sigma ** 2)) * np.exp(- ((x - (size-1)/2)**2 + (y - (size-1)/2)**2) / (2 * sigma ** 2)),
        (size, size)
    )
    return kernel
kernel_size = 3
sigma = 1.5
# Tạo kernel bộ lọc Gaussian
gaussian_filter = gaussian_kernel(kernel_size, sigma)

print("Kernel bộ lọc Gaussian:")
print(gaussian_filter)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('./flower.jpg')

image_float = image.astype(float)
kernel_size = 3
sigma = 1.4

kernel = cv2.getGaussianKernel(kernel_size, sigma)
gaussian_filter = np.outer(kernel, kernel.transpose())

#Làm mờ
blurred_image = cv2.filter2D(image_float, -1, gaussian_filter)
blurred_image = np.clip(blurred_image, 0, 255).astype(np.uint8)

plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(122), plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)), plt.title('Blurred Image')
plt.show()

```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve

input = './output_image.bmp'
output = './output.bmp'

image = Image.open(input)
image_array = np.array(image)

# print(image_array[0,1,0])
# print(image_array[0,1,1])
# print(image_array[0,1,2])

if len(image_array.shape) == 3:
     image_array = np.dot(image_array[..., :3], [0.299, 0.587, 0.114])

sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
#s_z = np.array([[0,1,1],[-1,0,1],[-1,-1,0]])

g_x = convolve(image_array, sobel_x)
g_y = convolve(image_array, sobel_y)

g_magnitude = np.sqrt(g_x**2 + g_y**2).astype(np.uint8)

output_image = Image.fromarray(g_magnitude)
output_image.save(output)

plt.subplot(121), plt.imshow(image_array , cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(g_magnitude, cmap='gray'), plt.title('Sobel Filter')
plt.show()
```

```python
file_path = "./out.txt"

with open(file_path, "r") as file:
    lines = file.readlines()

content_array = []
for line in lines:
    # Bỏ qua các dòng có dấu '//'
    if '//' in line:
        continue
    content_array.append(line.strip())

# Chuyển đổi phần tử hex thành decimal
binary_list = []
for i in range(118604):
    if content_array[i]:
        binary_element = int(content_array[i], 16)
        binary_list.append(binary_element)
    else:
        continue

j = 0
output_file_path = "./out_123456789.txt"
with open(output_file_path, "w") as output_file:
    for binary_element in binary_list:
        output_file.write(str(binary_element) + " ")
        j = j + 1
        if j == 298:
            output_file.write("\n")
            j = 0
```
